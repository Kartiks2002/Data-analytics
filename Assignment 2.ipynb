Assignment:

For the SP500 oil gold bitcoin.csv dataset:
(1) Find the correlation and covariance between BITCOIN, Gold, and S&P500. Print the correlation matrix and covariance matrix.
(2) Draw scatter plots and heatmaps for bitcoin-gold and bitcoin-s&p500 combinations.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('SP500 oil gold bitcoin.csv')
df.head()
```

```python
df.describe().transpose()
```

```python
type(df['Date'])
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
selected_columns = ["BITCOIN", "Gold", "S&P500"]
df_selected = df[selected_columns]
```

```python
# Calculating the correlation matrix
corr_matrix = df_selected.corr()
print("Correlation Matrix:")
print(corr_matrix)
```

#### Bitcoin and S&P 500 are highly correlated (0.86), which might suggest that Bitcoin is moving in a similar direction to the broader stock market (S&P 500).
#### Bitcoin and Gold have a moderate positive correlation (0.59), meaning they sometimes move together but not always.
#### Gold and S&P 500 are somewhat correlated (0.45), indicating a mild relationship between the movements of these two assets.

```python
# Calculating the covariance matrix
covariance_matrix = df_selected.cov()
print("Covariance Matrix:")
print(covariance_matrix)
```

#### Bitcoin and S&P 500 have a strong positive covariance, meaning they tend to move together.
#### Bitcoin and Gold have a moderate positive covariance.
#### Gold and S&P 500 have a much weaker positive covariance.

```python
# Scatter plot for Bitcoin vs Gold
plt.figure(figsize=(6, 4))
plt.scatter(df['BITCOIN'], df['Gold'], color='blue', alpha=0.7)
plt.title('Bitcoin vs Gold')
plt.xlabel('Bitcoin')
plt.ylabel('Gold')
plt.show()

# Scatter plot for Bitcoin vs S&P500
plt.figure(figsize=(6, 4))
plt.scatter(df['BITCOIN'], df['S&P500'], color='green', alpha=0.7)
plt.title('Bitcoin vs S&P500')
plt.xlabel('Bitcoin')
plt.ylabel('S&P500')
plt.show()

# Heatmap for Bitcoin vs Gold
plt.figure(figsize=(6, 4))
sns.heatmap(df_selected[['BITCOIN', 'Gold']].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Heatmap: Bitcoin vs Gold')
plt.show()

# Heatmap for Bitcoin vs S&P500
plt.figure(figsize=(6, 4))
sns.heatmap(df_selected[['BITCOIN', 'S&P500']].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Heatmap: Bitcoin vs S&P500')
plt.show()
```
